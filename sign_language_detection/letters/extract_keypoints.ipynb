```python
dir_path = os.path.join("..","..","data","letters","all_letters_2")
mp_holistic = mp.solutions.holistic
mp_drawing = mp.solutions.drawing_utils
holistic = mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5)


labels_dic = {
    "aleff": 0, #أ
    "bb": 1,    #ب
    "taa": 2,   #ت
    "thaa": 3,  #ث
    "jeem": 4,  #ج
    "haa": 5,   #ح
    "khaa": 6,  #خ
    "dal": 7,   #د
    "thal": 8,  #ذ
    "ra": 9,    #ر
    "zay": 10,  #ز
    "seen": 11, #س
    "sheen": 12,#ش
    "saad": 13, #ص
    "dhad": 14, #ض
    "ta": 15,   #ط
    "dha": 16,  #ظ
    "ain": 17,  #ع
    "ghain": 18,#غ
    "fa": 19,   #ف
    "gaaf": 20, #ق
    "kaaf": 21, #ك
    "laam": 22, #ل
    "meem": 23, #م
    "nun": 24,  #ن
    "ha": 25,   #هـ
    "waw": 26,  #و
    "ya": 27,   #ئ
    "toot": 28, #ة
    "al": 29,   #ال
    "la": 30,   #لا
    "yaa": 31   #ي
}




def mediapipe_detection(image,model):
    image  = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    image.flags.writeable = False
    results = model.process(image)
    image.flags.writeable = True
    image  = cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
    return image,results


def draw_styled_landmarks(image,results):
    # Draw pose connections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                             ) 
    # Draw left hand connections
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                             ) 
    # Draw right hand connections  
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS, 
                             mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4), 
                             mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                             ) 
    
```

```python
class ImageProcessor:
    def __init__(self,transformer=None):
        self.transformer = transformer # the datagenerator class
        self.seed = random.randint(1,100000000)
    
    
    def change_seed(self):
        self.seed = random.randint(1,100000000)
    
    def transform(self,frame):
        self.change_seed()
        lis = []
        for trans_frame in self.transformer.flow(np.expand_dims(frame, axis=0),seed=self.seed):
            lis.append(np.squeeze(trans_frame.astype(np.uint8), axis=0))
            if(len(lis)==2):break
        #return np.squeeze(trans_frame.astype(np.uint8), axis=0)
        return np.array(lis,dtype=np.uint8)
        

datagen = ImageDataGenerator(
        rotation_range=7,
        width_shift_range=0.13,
        height_shift_range=0.13,
        shear_range=0.14,
        zoom_range=0.14,
        #horizontal_flip=True,
        fill_mode='nearest')

image_processor = ImageProcessor(transformer=datagen)
```

```python
data   = [] 
labels = []

letters = os.listdir(dir_path)
finish = False
for letter in letters:
    images_path = os.path.join(dir_path,letter,'right')
    images = os.listdir(images_path)
    print(f"letter {letter} started")
    for image_num,image in enumerate(images):
        image_path = os.path.join(images_path,image)
        frame = cv2.imread(image_path)

        transformed_frames = image_processor.transform(frame)
        for i,transofrmed_frame in enumerate(transformed_frames):
            transofrmed_frame,results = mediapipe_detection(transofrmed_frame , holistic)
            draw_styled_landmarks(transofrmed_frame,results)
            if (results.right_hand_landmarks):
                points = results.right_hand_landmarks.landmark
                data.append(points)
                labels.append(labels_dic[letter])
        print(f"{image_num}/{len(images)}",end="\r")
    print(f"letter {letter} finished")


cv2.destroyAllWindows()

    
```
